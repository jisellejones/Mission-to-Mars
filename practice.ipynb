```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the quotes to Scrape site
url = 'http://quotes.toscrape.com'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, "html.parser")
```

```python
# Scrape the title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top 10 tags
tag_box = html_soup.find('div', class_='tags-box')

# tag box
tags = tag_box.find_all('a', class_="tag")

for tag in tags:
    word = tag.text
    print(word)
```

```python
# Create a for loop that will
# 1. Create a BeautifulSoup object
# 2. Find all the quotes on the page
# 3. Print each quote from the page
# 4. Click the "Next" button at the bottom of the page

for x in range(1, 6):
    html = browser.html
    quote_soup = soup(html, "html.parser")
    quotes = html_soup.find_all('span', class_="text")
    for quote in quotes:
        print('page', x, '----------')
        print(quote.text)
    browser.links.find_by_partial_text('Next').click()
```

```python
# Visit the books to Scrape site
url = 'http://books.toscrape.com'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, "html.parser")
```

```python
# Scrape the list of books on the left hand side of page

tag_box = html_soup.find('ul', class_='nav nav-list')
tags = tag_box.find_all('a')

for tag in tags:
    book = tag.text
    print(book)
```
